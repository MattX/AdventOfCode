```python
def follow(coords):
    pos = []
    x = 0
    y = 0
    for change in coords.split(','):
        pos.append((x, y))
        dir, *dist = change
        dist = int(''.join(dist))
        if dir == 'R':
            y += dist
        elif dir == 'U':
            x -= dist
        elif dir == 'L':
            y -= dist
        elif dir == 'D':
            x += dist
        else:
            raise RuntimeException()
    return min(p[0] for p in pos), max(p[0] for p in pos), min(p[1] for p in pos), max(p[1] for p in pos)
```

```python
print(follow(w1), follow(w2))
```

```python
import numpy as np
```

```python
def follow2(coords, grid):
    x = 10_000
    y = 10_000
    total_dist = 0
    for change in coords.split(','):
        dir = change[0]
        dist = int(change[1:])
        if dir == 'R':
            grid[x, y+1:y+dist+1] += np.asarray(range(1, dist+1)) + total_dist
            y += dist
        elif dir == 'U':
            grid[x-dist:x, y] += np.asarray(range(dist, 0, -1)) + total_dist
            x -= dist
        elif dir == 'L':
            grid[x, y-dist:y] += np.asarray(range(dist, 0, -1)) + total_dist
            y -= dist
        elif dir == 'D':
            grid[x+1:x+dist+1, y] += np.asarray(range(1, dist+1)) + total_dist
            x += dist
        else:
            raise RuntimeException()
        total_dist += dist
```

```python
grid1 = np.zeros((20_000, 20_000))
grid2 = np.zeros((20_000, 20_000))
```

```python
follow2(w1, grid1)
```

```python
follow2(w2, grid2)
```

```python
candidates = np.argwhere(np.logical_and(grid1 > 0, grid2 > 0))
len(candidates)
```

```python
def dist(pt):
    return grid1[pt[0], pt[1]] + grid2[pt[0], pt[1]]
```

```python
candidates_pos = [(c, dist(c)) for c in candidates]
```
